```python
import pandas as pd
import numpy as nd
```

```python
#dataset
df = pd.read_csv('Combined_News_DJIA.csv')
```

```python
df.shape
```

```python
# separating dataset...  may use regular train/test split later 
#***  not sure if i should be splitting the data by dates

train = df[df['Date'] < '2015-01-01']
test = df[df['Date'] > '2014-12-31']
#print train
#print test
train.shape
#test.shape
```

```python
# consolidate all To25 headlines into one chunk of strings for each row/date
train_headlines = []
for row in range(0, len(train.index)):
    train_headlines.append(''.join(str(x)for x in train.iloc[row, 2:27]))
```

```python
# import and instantiate CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
# learn the vocabulary of the training data, train_headlines
vect.fit(train_headlines)
```

```python
# examine the fitted vocabulary
vect.get_feature_names()
```

```python
# transform training data into sparse matrix
train_headline_spm = vect.transform(train_headlines)
train_headline_spm
```

```python
# convert sparse matrix to a dense matrix
train_headline_spm.toarray()
```

```python
train_headline_spm.shape
```

```python
# examine the vocabulary and document term matrix together
pd.DataFrame(train_headline_spm.toarray(), columns=vect.get_feature_names())
```

```python
# print sparse matrix
print(train_headline_spm)
```

```python
# 1st Model - Logistic Regression

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logregbase = logreg.fit(train_headline_spm, train['Label'])
```

```python
# consolidate test dataset

test_headlines = []
for row in range(0, len(test.index)):
    test_headlines.append(''.join(str(x) for x in test.iloc[row, 2:27]))
    
test_headline_spm = vect.transform(test_headlines)
test_headline_spm.shape
predictions = logregbase.predict(test_headline_spm)
```
